# Polynomial Regression 

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the Dataset

```python
dataset = pd.read_csv('Fish.csv')
```

```python
dataset.info()
```

```python
dataset.head()
```

## Taking care of Missing data

```python
dataset.isna().sum()
```

```python
x = dataset.drop(['Species','Weight'],axis = 1)
y = dataset['Weight']
```

## Training the Linear Regression model on the Whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Visualising the Linear Regression Result

```python
for i in x:
    sns.regplot(x = i,y = 'Weight',data = dataset)
    plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

## Model Score with Linear Regression

```python
lin_reg.score(x,y)
```

## Predicting a new result with Linear Regression

```python
lin_reg.predict([[10,10,10,10,10]])
```

## Training the Polynomial Regression model on the Whole dataset (degree = 2)

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg1 = PolynomialFeatures(degree = 2)
x_poly1 = poly_reg1.fit_transform(x)
lin_reg_poly1 = LinearRegression()
lin_reg_poly1.fit(x_poly1,y)
```

## Visualising the Polynomial Regression Result

```python
for i in x:
    plt.scatter(dataset[i],y,color='red')
    plt.plot(dataset[i],lin_reg_poly1.predict(x_poly1))
    plt.xlabel(i)
    plt.ylabel('Weigth')
    plt.title('(Using Degree = 2)')
    plt.show()
```

```python
plt.plot(x,lin_reg_poly1.predict(x_poly1))
plt.show()
```

## Model Score with Polynomial  Regression

```python
lin_reg_poly1.score(x_poly1,y)
```

## Predicting a new result with Polynomial Regression

```python
lin_reg_poly1.predict(poly_reg1.fit_transform([[10,10,10,10,10]]))
```

## Training the Polynomial Regression model on the Whole dataset (degree = 10)

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2 = PolynomialFeatures(degree = 10)
x_poly2 = poly_reg2.fit_transform(x)
lin_reg_poly2 = LinearRegression()
lin_reg_poly2.fit(x_poly2,y)
```

## Visualising the Polynomial Regression Result

```python
for i in x:
    plt.scatter(dataset[i],y,color='red')
    plt.plot(dataset[i],lin_reg_poly2.predict(x_poly2))
    plt.xlabel(i)
    plt.ylabel('Weigth')
    plt.title('(Using Degree = 10)')
    plt.show()
```

```python
plt.plot(x,lin_reg_poly2.predict(x_poly2))
plt.show()
```

## Model Score with Polynomial Regression

```python
lin_reg_poly2.score(x_poly2,y)
```

## Predicting a new result with Polynomial Regression

```python
lin_reg_poly2.predict(poly_reg2.fit_transform([[10,10,10,10,10]]))
```
